```python
#Python Guess Game by Importing random Library
#Importing the random library to choose random word from the given list
import random 
name_person = input("What is your name? ") 

# Here we  asked to enter the name first 

print("All the Best ! ", name_person) 

words = ['python', 'java', 'c++', 'programming','Machine Learning', 'AI', 'Regression', 'ifelse', 'lockdown', 'strings', 'loops', 'jobs'] 

# Function will choose one of the random 
# word from this list of words 
word = random.choice(words) 
print("Guess the characters") 
guesses = '' 

# we are taking 15 turns as we can take any number of turns
turns = 15


while turns > 0: 
	
	# counts the number of times as user fails 
	failed = 0
	
	# all characters from the input 
	for char in word: 
		
		# comparing that character with 
		# the character in guesses 
		if char in guesses: 
			print(char) 
			
		else: 
			print("_") 
			
			# for every failure 1 will be 
			# incremented in failure 
			failed += 1
			

	if failed == 0: 
		# user will win the game if failure is 0 
		# 'You Win' will be given as output 
		print("You Win") 
		
		# this print the correct word 
		print("The word is: ", word) 
		break
	
	# if user has input the wrong alphabet then 
	# it will ask user to enter another alphabet 
	guess = input("guess a character:") 
	
	# every input character will be stored in guesses 
	guesses += guess 
	
	# check input with the character in word 
	if guess not in word: 
		
		turns -= 1
		
		# if the character doesn’t match the word 
		# then “Wrong” will be given as output 
		print("Wrong") 
		
		# this will print the number of 
		# turns left for the user 
		print("You have", + turns, 'more guesses') 
		
		
		if turns == 0: 
			print("You Loose") 
```
